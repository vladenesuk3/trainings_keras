```python
import pandas as pd
import numpy as np

np.set_printoptions(precision=3, suppress=True)

import tensorflow as tf
from tensorflow import keras
```

```python
abalone_train = pd.read_csv(
    "https://storage.googleapis.com/download.tensorflow.org/data/abalone_train.csv",
    names=["Length", "Diameter", "Height", "Whole weight", "Shucked weight",
           "Viscera weight", "Shell weight", "Age"])

abalone_train
```

```python
abalone_train.describe().transpose()
```

```python
abalone_train_copy = abalone_train.copy()
abalone_train_labels = abalone_train.pop('Age')
```

```python
abalone_features = np.array(abalone_train_copy)
```

```python
abalone_train.shape, abalone_train_labels.shape
```

```python
model_test = tf.keras.Sequential([
    keras.layers.Dense(units=64, activation='sigmoid', input_shape=(7,)),
    keras.layers.Dense(units=128, activation='sigmoid'),
    keras.layers.Dense(units=1)
])

model_test.compile(optimizer=keras.optimizers.SGD(learning_rate=0.1), loss=keras.losses.mean_absolute_error, metrics='mse')
model_test.fit(abalone_train, abalone_train_labels, epochs=100, validation_split=0.2)
```

```python
import keras_tuner as kt
import os
import shutil
```

```python
def build_reg(hp):
    model = tf.keras.Sequential()
    
    hp_units = hp.Int('units', min_value=32, max_value=512, step=32)
    hp_activation = hp.Choice('activation', values=['relu', 'elu', 'tanh'])
    
    model.add(keras.layers.Dense(units=hp_units, activation=hp_activation))
    model.add(keras.layers.Dense(units=hp_units, activation=hp_activation))
    model.add(keras.layers.Dense(1))
    
    hp_learning_rate = hp.Choice('learning_rate', values=[0.1, 0.01, 0.001])
    model.compile(optimizer=keras.optimizers.SGD(learning_rate=hp_learning_rate), 
                  loss=tf.keras.losses.mean_absolute_error, metrics=['mse', 'mae'])
    
    return model
```

```python
tuner = kt.Hyperband(build_reg, objective='val_mse', max_epochs=10, factor=3, 
                     directory='my_dir', project_name='kt_test_1')
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5)
```

```python
tuner.search(abalone_train, abalone_train_labels, epochs=50, validation_split=0.2, callbacks=[early_stop])
```

```python
best_kt = tuner.get_best_hyperparameters(num_trials=1)[0]
```

```python
logdir = "logs/tuner/" + 'model_tuner_regr'
tensorboard_callback = keras.callbacks.TensorBoard(log_dir=logdir)

model_tuner = tuner.hypermodel.build(best_kt)
history_tuner = model_tuner.fit(abalone_train, abalone_train_labels, epochs=500, validation_split=0.2, callbacks=[tensorboard_callback])
```
